```python
import pandas as pd
import prince
import numpy
```

```python
df = pd.read_csv('CSV/Terry_Stops.csv')
```

```python
df = df[['Frisk Flag', 
             'Officer Race', 
             'Officer Gender', 
             'Subject Perceived Race', 
             'Subject Age Group', 
             'Subject Perceived Gender']].copy()
```

```python
print("The original data set contains: {} rows and {} columns".format(df.shape[0], df.shape[1]))
print("Features of the original data set:\n", list(df.columns))
```

```python
# number of categorical features
print("Number of categorical features : {}".format(len(df.select_dtypes(include=['object']).columns)))
print("Number of continuous features : {}".format(len(df.select_dtypes(include=['int64', 'float64']).columns)))
```

```python
df.nunique()
```

```python
mca = prince.MCA(
        n_components=6,
        n_iter=100,
        copy=True,
        check_input=True,
        engine='auto',
        random_state=42
        )
df_mca = mca.fit(df)

ax = df_mca.plot_coordinates(
        X=df,
        ax=None,
        figsize=(8, 10),
        show_row_points=False,
        row_points_size=0,
        show_row_labels=False,
        show_column_points=True,
        column_points_size=30,
        show_column_labels=True,
        legend_n_cols=1
               ).legend(loc='center left', bbox_to_anchor=(1, 0.5))
```
